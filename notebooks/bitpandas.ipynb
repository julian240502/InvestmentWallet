```python
import os
from dotenv import load_dotenv

# Charger le contenu du fichier .env dans l'environnement
load_dotenv()

# Récupérer la clé API bitpandas depuis la variable d'environnement
BITPANDAS_API_KEY = os.getenv('BITPANDAS_API_KEY')
```

```python
import requests

# Exemple d'URL pour récupérer le ticker (vérifiez l'URL exacte dans la documentation)
url = "https://api.bitpanda.com/v1/ticker"

# Envoyer la requête GET sans authentification (pour l’API publique)
response = requests.get(url)

if response.ok:
    data = response.json()
    print("Données du ticker :", data)
else:
    print("Erreur lors de la récupération des données :", response.status_code)
```

```python
import os
import requests

url = "https://api.bitpanda.com/v1/trades"

headers = {
    'X-Api-Key': BITPANDAS_API_KEY  
}

response = requests.get(url, headers=headers)

if response.ok:
    account_data = response.json()
    print("Données du compte :", account_data)
else:
    print("Erreur :", response.status_code)
```

```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
from src.ingestion import enrich_wallet_with_price

from src.user import get_wallets , get_trades, get_wallets_transactions
df2 = get_wallets(process=True)

df_wallets = enrich_wallet_with_price(df2)
df_wallets.head(20)
```

```python
import matplotlib.pyplot as plt
# On filtre pour ne garder que les actifs pour lesquels le total_value est défini
df_valid = df_wallets.dropna(subset=["total_value"])
df_valid = df_valid[df_valid["total_value"] > 0]

# Agrégation par symbole (au cas où vous auriez plusieurs lignes par symbole)
aggregated = df_valid.groupby("symbol")["total_value"].sum()

# Création du graphique en camembert
plt.figure(figsize=(8, 6))
plt.pie(aggregated, labels=aggregated.index, autopct='%1.1f%%', startangle=90)
plt.title("Répartition du portefeuille")
plt.show()
```
