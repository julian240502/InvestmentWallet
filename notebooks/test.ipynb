```python
# Charger les etfs
nasdaq_100 = yf.Ticker("ANX.PA")
eurostoxx_50 = yf.Ticker("E50.PA")
sp500 = yf.Ticker("ES5.PA")

# Charger les actions
dassault_systemes =yf.Ticker("DSY.PA")
microsoft = yf.Ticker("MSFT")
nvidia = yf.Ticker("NVDA")

# Charger les cryptos
BTC = yf.Ticker("BTC-EUR")
SOL = yf.Ticker("SOL-EUR")
XRP = yf.Ticker("XRP-EUR")
ETH = yf.Ticker("ETH-EUR")


etfs = [nasdaq_100, eurostoxx_50, sp500]
cryptos = [BTC, SOL, XRP, ETH]
actions = [microsoft, dassault_systemes, nvidia]
```

```python
import yfinance as yf

# Définir les symboles par catégorie
etf_symbols = ["ANX.PA", "E50.PA", "ES5.PA"]
action_symbols = ["MSFT", "DSY.PA", "NVDA"]
crypto_symbols = ["BTC-EUR", "SOL-EUR", "XRP-EUR", "ETH-EUR"]

# Définir la période souhaitée
start_date = "2020-01-01"
end_date = "2025-01-01"

# Télécharger les données pour chaque catégorie
data_etfs = yf.download(etf_symbols, start=start_date, end=end_date, group_by='ticker')
data_actions = yf.download(action_symbols, start=start_date, end=end_date, group_by='ticker')
data_cryptos = yf.download(crypto_symbols, start=start_date, end=end_date, group_by='ticker')

# Vous pouvez ensuite sauvegarder les résultats pour un accès ultérieur :
data_etfs.to_csv("../data/raw/etfs.csv")
data_actions.to_csv("../data/raw/actions.csv")
data_cryptos.to_csv("../data/raw/cryptos.csv")
```
